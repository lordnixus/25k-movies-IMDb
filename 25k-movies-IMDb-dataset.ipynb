```python
def findmovie(df,query,score):
  filter1=df['movie title'].str.contains(query.title())
  filter2=df['Overview'].str.contains(query)
  filter3=df['Rating']>score
  df=df[(filter1 | filter2) & filter3]
  return df
```

```python
findmovie(df,"war",7.5)
```

```python
findmovie(df,"space",7.0)
```
